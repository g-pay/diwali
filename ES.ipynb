```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn import preprocessing
```

```python
X=df.iloc[:, :df.shape[1]-1]       #Independent Variables
y=df.iloc[:, -1]                   #Dependent Variable
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=8)
```

```python
models = {
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=2),
    "Linear SVM":LinearSVC(random_state=8, max_iter=900000),
    "Polynomical SVM":SVC(kernel="poly", degree=2, random_state=8),
    "RBF SVM":SVC(kernel="rbf", random_state=8),
    "Sigmoid SVM":SVC(kernel="sigmoid", random_state=8)
}
```

```python
from sklearn.impute import SimpleImputer

# Instantiate an imputer
imputer = SimpleImputer(strategy='mean')

# Fit and transform the imputer on your training data
X_train = imputer.fit_transform(X_train)

# Transform your test data using the same imputer
X_test = imputer.transform(X_test)
```

```python
df.fillna(0, inplace=True)
```

```python
for model_name, model in models.items():
    y_pred=model.fit(X_train, y_train).predict(X_test)
    print(f"Accuracy for {model_name} model \t: {metrics.accuracy_score(y_test, y_pred)}")
```

```python
from sklearn.metrics import classification_report

# Create an empty dictionary to store classification reports
classification_reports = {}

# Iterate through the models and compute classification reports
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred)
    classification_reports[model_name] = report

# Display the classification reports
for model_name, report in classification_reports.items():
    print(f"Classification Report for {model_name}:")
    print(report)
    print()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

models = {
    "K-Nearest Neighbors": KNeighborsClassifier(n_neighbors=2),
    "Linear SVM": LinearSVC(random_state=8, max_iter=900000),
    "Polynomial SVM": SVC(kernel="poly", degree=2, random_state=8),
    "RBF SVM": SVC(kernel="rbf", random_state=8),
    "Sigmoid SVM": SVC(kernel="sigmoid", random_state=8)
}

# Create an empty dictionary to store confusion matrices
confusion_matrices = {}

# Iterate through the models and compute confusion matrices
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    confusion_matrices[model_name] = cm

# Display the confusion matrices
for model_name, cm in confusion_matrices.items():
    print(f"Confusion Matrix for {model_name}:")
    print(cm)
    accuracy = accuracy_score(y_test, models[model_name].predict(X_test))
    print(f"Accuracy for {model_name}: {accuracy}")
    print()
```
